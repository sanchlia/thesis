```python
import os 
import pandas as pd 
import numpy as np
```

## Income

```python
name = "income"
base_path = "/Users/sauravanchlia/Fair_ML/Projects/thesis/results/income"
```

```python
cols = ['metric', 'LR', 'reduction', 'reweight']
df = []
for noise in ['bias', 'balanced', 'flip']:
    for level in ['0.1','0.3']:
        audit_df = pd.read_csv(os.path.join(base_path, f"{name}_{noise}_{level}", "audit.csv"))
        meta = [noise, level]
        for metric in ['accuracy', 'Equal_Opportunity','equalized_odds']:
            for eval_type in [ 'noisy','fair_clean', 'emp_clean']:
                temp = [metric, eval_type]
                for m in ['LR', 'Reduction', 'Reweight']:
                    temp_df = audit_df[(audit_df['model'] == m)]
                    # temp.append((temp_df[temp_df['eval_type'] == eval_type][metric].reset_index() - temp_df[temp_df['eval_type'] == 'ground'][metric].reset_index())[metric].mean())
                    temp.append(
                        f"{(temp_df[temp_df['eval_type'] == eval_type][metric].reset_index() - temp_df[temp_df['eval_type'] == 'ground'][metric].reset_index())[metric].mean():.3f} + {(temp_df[temp_df['eval_type'] == eval_type][metric].reset_index() - temp_df[temp_df['eval_type'] == 'ground'][metric].reset_index())[metric].std():.3f}"
                        )
                df.append( meta + temp)
        # print(audit_df)
        
```

```python
rdf = pd.DataFrame(df, columns=['noise', 'level', 'metric', 'eval_type', 'LR', "Reduction", "Reweight"])
```

```python
rdf
```

```python
for row in rdf.iterrows():
    str = []
    _n = "\\"
    if not row[1]["noise"] == 'balanced': continue
    if not row[1]['level'] == '0.3': continue
    print(f"{row[1]['metric']} & {row[1]['eval_type']}")
    for m in ['LR', 'Reduction', 'Reweight']:
        f,s = row[1][m].split("+")
        str.append(f"${f} {_n}pm {s}$ ")
    print(" & ".join(str))
    
    
```

```python
rdf
```

```python
rdf.to_csv('tables/audit_income.csv')
```

```python
sdf =  pd.DataFrame(df, columns=['noise', 'level', 'metric', 'eval_type', 'LR', "Reduction", "Reweight"])
```

```python
def print_table(df):
    for noise in ['balanced', 'bias', 'flip']:
        for level in ['0.1','0.3']:
            if noise != 'flip': continue
            temp = df[(df['noise']== noise) & (df['level'] == level)]
            print(f'{noise=} and {level=}')
            for row in temp.iterrows():
                str = []
                print(f"{row[1]['metric']} & {row[1]['eval_type']}")
                for m in ['LR', 'Reduction', 'Reweight']:
                    f,s = row[1][m].split("+")
                    str.append(f"${f} {_n}pm {s}$ ")
                print(" & ".join(str))
```

```python
print_table(sdf)
```
