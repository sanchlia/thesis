```python
def create_table(name, noises, levels, base_path, save_path):
    for noise in noises:
        for level in levels:
            df = []
            audit_df = pd.read_csv(os.path.join(base_path, f"{name}_{noise}_{level}", "audit.csv"))
            fair_cleanup = pd.read_csv(os.path.join(base_path, f"{name}_{noise}_{level}", "fair_cleanup.csv"))
            emp_cleanup = pd.read_csv(os.path.join(base_path, f"{name}_{noise}_{level}", "emp_cleanup.csv"))
            fair_reduction = pd.read_csv(os.path.join(base_path, f"{name}_{noise}_{level}","fair_reduction_2.csv"))
            # meta = []

            for metric in ['accuracy', 'Equal_Opportunity', 'equalized_odds']:
                for train_type in ['noisy', 'fair', 'emp']:
                    meta =  [noise, level, metric]
                    model_list = []
                    for m in ['LR', 'Reweight', 'Reduction']:
                        row = []
                        for eval_type in ['ground']:
                            if train_type == 'noisy':
                                # read from audit

                                row.append(
                                    f"{audit_df[(audit_df['model'] == m) & (audit_df['eval_type'] == eval_type)][metric].mean()} + {audit_df[(audit_df['model'] == m) & (audit_df['eval_type'] == eval_type)][metric].std()} "
                                    )

                            elif train_type == 'fair':
                                row.append(
                                    f"{fair_cleanup[(fair_cleanup['model'] == m) & (fair_cleanup['eval_type'] == eval_type)][metric].mean()} + {fair_cleanup[(fair_cleanup['model'] == m) & (fair_cleanup['eval_type'] == eval_type)][metric].std()}"
                                    )
                            elif train_type == 'emp':
                                row.append(
                                    f"{emp_cleanup[(emp_cleanup['model'] == m) & (emp_cleanup['eval_type'] == eval_type)][metric].mean()} +  {emp_cleanup[(emp_cleanup['model'] == m) & (emp_cleanup['eval_type'] == eval_type)][metric].std()}"  
                                    )
                        model_list.append(meta + [f"{m}_{train_type}"] +  row)
                    # add fair reduction here 
                    for w in ['fair', 'emp']:
                        model_list.append(meta + [f'FR_{w}'] + 
                            [f"{fair_reduction[(fair_reduction['weights'] == w) & (fair_reduction['eval_type'] == 'ground')][metric].mean()} + {fair_reduction[(fair_reduction['weights'] == w) & (fair_reduction['eval_type'] == 'ground')][metric].std()}"]
                        )
                    df.append(model_list)
            ndf = []
            for item in df:
                for sitem in item:
                    ndf.append(sitem)

            adf = pd.DataFrame(ndf, columns=['noise', 'level', 'metric',  'model', 'ground'])
            os.makedirs(os.path.join(save_path, name, f"{name}_{noise}_{level}"), exist_ok=True)
            # adf.to_csv(os.path.join(save_path, name, f"{name}_{noise}_{level}", "fair_learn.csv"))
    return adf
                
                
```

```python
def create_latex_lines(name):
    bal_1_df = pd.read_csv(os.path.join(save_path, name, f"{name}_balanced_0.1", "fair_learn.csv"))
    bal_3_df = pd.read_csv(os.path.join(save_path, name, f"{name}_balanced_0.3", "fair_learn.csv"))
    b_1_df = pd.read_csv(os.path.join(save_path, name, f"{name}_bias_0.1", "fair_learn.csv"))
    b_3_df = pd.read_csv(os.path.join(save_path, name, f"{name}_bias_0.3", "fair_learn.csv"))
    f_1_df = pd.read_csv(os.path.join(save_path, name, f"{name}_flip_0.1", "fair_learn.csv"))
    f_3_df = pd.read_csv(os.path.join(save_path, name, f"{name}_flip_0.3", "fair_learn.csv"))
    
    content = []
    for idx in range(len(bal_1_df)):
        if bal_1_df.iloc[idx]['metric'] == 'Equal_Opportunity': continue
        content.append(
            f"{bal_1_df.iloc[idx]['metric']} & {bal_1_df.iloc[idx]['model']} & ${bal_1_df.iloc[idx]['ground'].split('+')[0]} \pm {bal_1_df.iloc[idx]['ground'].split('+')[1]}$ & ${bal_3_df.iloc[idx]['ground'].split('+')[0]} \pm {bal_3_df.iloc[idx]['ground'].split('+')[1]}$  & ${b_1_df.iloc[idx]['ground'].split('+')[0]} \pm {b_1_df.iloc[idx]['ground'].split('+')[1]}$ & ${b_3_df.iloc[idx]['ground'].split('+')[0]} \pm {b_3_df.iloc[idx]['ground'].split('+')[1]}$ & ${f_1_df.iloc[idx]['ground'].split('+')[0]} \pm {f_1_df.iloc[idx]['ground'].split('+')[1]}$ & ${f_3_df.iloc[idx]['ground'].split('+')[0]} \pm {f_3_df.iloc[idx]['ground'].split('+')[1]}$"
        )
    return content
```
